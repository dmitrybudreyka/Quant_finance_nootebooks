<center> # Geometric Brownian Motion <center>

## continuous time

$$
\frac{dS_t}{S_t} = \mu dt + \sigma dW_t
$$

## discrete time

$$
\frac{S_{t+dt} - S_t}{S_t}
= \mu dt + \sigma \sqrt{dt}\xi_t
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def gbm(n_years=10, n_scenarios=50, n_steps_per_year=12, mu=0.07, sigma=0.15, s_0=100):
    dt = 1 / n_steps_per_year
    n_steps = n_years * n_steps_per_year
    z = np.random.standard_normal(size=(n_steps, n_scenarios))
    rets = mu * dt + sigma * np.sqrt(dt) * z
    rets = pd.DataFrame(rets)
    rets = pd.concat([pd.DataFrame(np.zeros(n_scenarios)).T, rets], ignore_index=True)
    return (s_0 * (1 + rets).cumprod())
```
