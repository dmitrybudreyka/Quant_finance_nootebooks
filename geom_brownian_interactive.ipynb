# Geometric Brownian Motion

## continuous time

$$
\frac{dS_t}{S_t} = \mu dt + \sigma dW_t
$$

## discrete time

$$
\frac{S_{t+dt} - S_t}{S_t}
= \mu dt + \sigma \sqrt{dt}\xi_t
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def gbm(n_years=10, n_steps_per_year=12, mu=0.07, sigma=0.15):
    dt = 1 / n_steps_per_year
    n_steps = n_years * n_steps_per_year
    z = np.random.standard_normal(size=n_steps)
    rets = mu * dt + sigma * np.sqrt(dt) * z
    return pd.DataFrame(rets)
    
```

```python
rets = gbm()
wealth = (1 + rets).cumprod()
wealth.plot()
```
